```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('csv-data\\cleaned_dataset_analysis.csv')
```

```python
plt.figure(figsize = [10, 8])
#plt.style.use('ggplot')
sns.set_style('whitegrid')
#sns.histplot(df, x = 'Price')
#sns.histplot(df, x = 'Price', binwidth=50000)
#sns.histplot(df, x = 'Price', bins=20)
sns.histplot(df, x='Price', bins=30, kde=True)
#sns.histplot(df, x = 'Price', log_scale=True)
plt.title("Distribution of Property Prices", fontsize = 16, fontweight = 'bold')
plt.xlabel('Price (€)', fontsize = 14, fontweight = 'bold', loc = 'right')
plt.ylabel('Count', fontsize = 14, fontweight = 'bold', loc = 'top')
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)
plt.show()
```

```python
plt.figure(figsize = [10, 8])
sns.set_style('whitegrid')
sns.boxplot(df, x = 'Price', flierprops={'marker': 'o', 'markerfacecolor': 'red', 'markersize': 8})
#plt.xscale('log')
q1, q2, q3 = df['Price'].quantile([0.25, 0.5, 0.75])
plt.axvline(q1, color='blue', linestyle='--', label='Q1')
plt.axvline(q2, color='green', linestyle='-', label='Median')
plt.axvline(q3, color='red', linestyle='--', label='Q3')
plt.legend()

plt.title('Statistics of Property Prices', fontsize = 16, fontweight = 'bold')
plt.xlabel('Price (€)', fontsize = 14, fontweight = 'bold', loc = 'right')
plt.ylabel('Value', fontsize = 14, fontweight = 'bold', loc = 'top')
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)
plt.show()
```
