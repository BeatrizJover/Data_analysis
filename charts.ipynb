```python
#Calculate quartiles and IQR
Q1 = df['Price'].quantile(0.25)  # (25%)
Q3 = df['Price'].quantile(0.75)  # (75%)
IQR = Q3 - Q1  

# Calculate limits for outliers
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

no_outliers_price = df[(df['Price'] > lower) & (df['Price'] < upper)]
print(f"Lower limit: {lower}")
print(f"Upper limit: {upper}")
outliers = df[(df['Price'] < lower) | (df['Price'] > upper)]
print("\nOutliers:")
print(len(outliers))
print(outliers)

print("\nNo Outliers:")
print(len(no_outliers_price))
print(no_outliers_price)
```

```python
no_outliers_space['Price per m2'] = round(no_outliers_space['Price'] / no_outliers_space['Livable Space (m2)'], 0)
```

```python
# Pivot the data
heatmap_data = no_outliers_space.pivot_table(
    values='Price per m2',  # The value to aggregate
    index='Subtype of Property',  # Rows
    columns='Furnished',  # Columns
    aggfunc='mean'  # Aggregation function
).sort_values(by=heatmap_data.columns[0], ascending=False)

# Create the heatmap
plt.figure(figsize=(8, 9))
sns.set_style('whitegrid')
sns.heatmap(
    heatmap_data,
    annot=True,  # Display values inside the cells
    fmt=".0f",  # Format for annotation (no decimal points)
    cmap='YlGnBu',  # Colormap for the heatmap
    linewidths=0.5  # Line thickness between cells
)

# Add titles and labels
plt.title("Average Price/m2 by Property Type and furniture availability", fontsize=16, fontweight='bold')
plt.xlabel("Furniture", fontsize=14, fontweight='bold')
plt.ylabel("Property Type", fontsize=14, fontweight='bold')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Pivot the data and reset index for better handling
barplot_data = no_outliers_space.pivot_table(
    values='Price per m2', 
    index='Subtype of Property', 
    columns='Furnished', 
    aggfunc='mean'
).reset_index()

# Melt the pivot table to make it long-format
barplot_data = barplot_data.melt(
    id_vars='Subtype of Property', 
    var_name='Furnished', 
    value_name='Price per m2'
)

# Sort the data by the average 'Price per m2' (descending order)
avg_prices = barplot_data.groupby('Subtype of Property')['Price per m2'].mean().sort_values(ascending=False).index
barplot_data['Subtype of Property'] = pd.Categorical(
    barplot_data['Subtype of Property'], 
    categories=avg_prices, 
    ordered=True
)

# Create the barplot
plt.figure(figsize=(12, 10))
sns.set_style('whitegrid')
sns.barplot(
    data=barplot_data, 
    x='Subtype of Property', 
    y='Price per m2', 
    hue='Furnished',  # Adds a legend for "Furnished" status
    palette='YlGnBu'
)

# Add titles and labels
plt.title("Average Price/m2 by Property Type and Furnished Status", fontsize=16, fontweight='bold')
plt.xlabel("Subtype of Property", fontsize=14, fontweight='bold')
plt.ylabel("Price per m2", fontsize=14, fontweight='bold')
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)

# Show the plot
plt.tight_layout()
plt.show()
```
