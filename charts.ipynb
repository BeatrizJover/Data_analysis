```python
plt.figure(figsize = [10, 8])
#plt.style.use('ggplot')
sns.set_style('whitegrid')
#sns.histplot(df, x = 'Price')
#sns.histplot(df, x = 'Price', binwidth=50000)
#sns.histplot(df, x = 'Price', bins=20)
sns.histplot(df, x='Price', bins=30, kde=True)
#sns.histplot(df, x = 'Price', log_scale=True)
plt.title("Distribution of Property Prices", fontsize = 16, fontweight = 'bold')
plt.xlabel('Price ()', fontsize = 14, fontweight = 'bold', loc = 'right')
plt.ylabel('Count', fontsize = 14, fontweight = 'bold', loc = 'top')
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)
plt.show()
```

```python
#Calculate quartiles and IQR
Q1 = df['Price'].quantile(0.25)  # (25%)
Q3 = df['Price'].quantile(0.75)  # (75%)
IQR = Q3 - Q1  

# Calculate limits for outliers
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

no_outliers_price = df[(df['Price'] > lower) & (df['Price'] < upper)]
print(f"Lower limit: {lower}")
print(f"Upper limit: {upper}")
outliers = df[(df['Price'] < lower) | (df['Price'] > upper)]
print("\nOutliers:")
print(len(outliers))
print(outliers)

print("\nNo Outliers:")
print(len(no_outliers_price))
print(no_outliers_price)
```

```python
#Calculate quartiles and IQR
Q1 = df['Livable Space (m2)'].quantile(0.25)  # (25%)
Q3 = df['Livable Space (m2)'].quantile(0.75)  # (75%)
IQR = Q3 - Q1  

# Calculate limits for outliers
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

no_outliers_space = no_outliers_price[(no_outliers_price['Livable Space (m2)'] > lower) & (df['Livable Space (m2)'] < upper)]
print(f"Lower limit: {lower}")
print(f"Upper limit: {upper}")
outliers = no_outliers_price[(no_outliers_price['Livable Space (m2)'] < lower) | (df['Livable Space (m2)'] > upper)]
print("\nOutliers:")
print(len(outliers))
print(outliers)

print("\nNo Outliers:")
print(len(no_outliers_space))
print(no_outliers_space)
```
